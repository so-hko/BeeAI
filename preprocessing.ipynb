```
# -*- coding: utf-8 -*-
"""DataFrame Example의 사본

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1_6UmCYnQ4DA4-oNuLaHJr8eZeiuGRWpn
"""

#from google.colab import drive
#drive.mount("/content/gdrive")

#import pandas as pd
#import numpy as np

'''
f = open("/content/gdrive/My Drive/data/watertreatment.data", "r")

Attributes = {"Day":[], "Q-E":[], "ZN-E":[], "PH-E":[], "DBO-E":[], "DQO-E":[],
              "SS-E":[], "SSV-E":[], "SED-E":[], "COND-E":[], "PH-P":[], "DBO-P":[],
              "SS-P":[], "SSV-P":[], "SED-P":[], "COND-P":[], "PH-D":[], "DBO-D":[],
              "DQO-D":[], "SS-D":[], "SSV-D":[], "SED-D":[], "COND-D":[], "PH-S":[],
              "DBO-S":[], "DQO-S":[], "SS-S":[], "SSV-S":[], "SED-S":[], "COND-S":[],
              "RD-DBO-P":[], "RD-SS-P":[], "RD-SED-P":[], "RD-DBO-S":[], "RD-DQO-S":[],
              "RD-DBO-G":[], "RD-DQO-G":[], "RD-SS-G":[], "RD-SED-G":[]}

Count_Wonder = {"Day":0, "Q-E":0, "ZN-E":0, "PH-E":0, "DBO-E":0, "DQO-E":0,
                "SS-E":0, "SSV-E":0, "SED-E":0, "COND-E":0, "PH-P":0, "DBO-P":0,
                "SS-P":0, "SSV-P":0, "SED-P":0, "COND-P":0, "PH-D":0, "DBO-D":0,
                "DQO-D":0, "SS-D":0, "SSV-D":0, "SED-D":0, "COND-D":0, "PH-S":0,
                "DBO-S":0, "DQO-S":0, "SS-S":0, "SSV-S":0, "SED-S":0, "COND-S":0,
                "RD-DBO-P":0, "RD-SS-P":0, "RD-SED-P":0, "RD-DBO-S":0, "RD-DQO-S":0,
                "RD-DBO-G":0, "RD-DQO-G":0, "RD-SS-G":0, "RD-SED-G":0}

'''

# Another dataset

import pandas as pd
import numpy as np
import torch.nn as nn
import torch
import torch.nn.functional as F
import random
from torch.utils.data import TensorDataset
from torch.utils.data import DataLoader

f = open("/Faults.NNA", "r")
Attributes = {"X_Minimum":[], "X_Maximum":[], "Y_Minimum":[], "Y_Maximum":[], "Pixels_Areas":[], "X_Perimeter":[], "Y_Perimeter":[],
              "Sum_of_Luminosity": [], "Minimum_of_Luminosity":[], "Maximum_of_Luminosity":[], "Length_of_Conveyer":[], "TypeOfSteel_A300":[],
              "TypeOfSteel_A400":[], "Steel_Plate_Thickness":[], "Edges_Index":[], "Empty_Index":[], "Square_Index":[], "Outside_X_Index":[],
              "Edges_X_Index":[], "Edges_Y_Index":[], "Outside_Global_Index":[], "LogOfAreas":[], "Log_X_Index":[], "Log_Y_Index":[], "Orientation_Index":[],
              "Luminosity_Index":[], "SigmoidOfAreas":[], "Pastry":[], "Z_Scratch":[], "K_Scatch":[], "Stains":[], "Dirtiness":[], "Bumps":[], "Other_Faults":[] }

use_cuda = torch.cuda.is_available()
DEVICE = torch.device("cuda" if use_cuda else "cpu")

dataset = []

Types = {}
flag = True

for i in range(len(Attributes)):
  Types[i] = []

for text in f:
  t = text.split()
  for i, Attr in enumerate(Attributes.keys()):
    if t[i].find(".") != -1:
      q = float(t[i])
      if flag == True:
        Types[i] += ["float"]

    else:
      q = int(t[i])
      if flag == True:
        Types[i] += ["int"]

    Attributes[Attr] += [q]
  flag = False

for i in range(len(Attributes["X_Minimum"])): # Dataset load
  save = []
  for Attr in Attributes.keys():
    save += [Attributes[Attr][i]]
  dataset += [save]

print(len(dataset))

x_train = []
y_train = []

for j in range(len(dataset)):
  x_train += [dataset[j][:-7]]

  p = dataset[j][-7:]
  idx = p.index(1)
  y_train += [[idx]]
  #y_train += [dataset[j][-7:]]

# 전체 데이터 shuffle(reason : 1, 2, 3, 4, 5, 6, 7 순으로 데이터 되어있음.)
# shuffle이 1차원 배열만을 섞는것을 이용
r = [[x, y] for x, y in zip(x_train, y_train)]
random.shuffle(r)

x_train = [x[0] for x in r]
y_train = [x[1] for x in r]


# data의 수 contract -- 1941
x_len = len(x_train[0])
y_len = len(y_train[0])

x_allData = torch.FloatTensor(x_train).to(DEVICE)
y_allData = torch.LongTensor(y_train).to(DEVICE)

x_train = x_allData[:1359]
y_train = y_allData[:1359]

x_test = x_allData[1359:1747]
y_test = y_allData[1359:1747]

x_valid = x_allData[1747:]
y_valid = y_allData[1747:]


dataSet = TensorDataset(x_train, y_train)
dataloader = DataLoader(dataSet, batch_size=32, shuffle=True) # 배치크기 32, + Data shuffle
dataloader2 = DataLoader(dataSet, batch_size=32, shuffle=True)

valid_dataSet = TensorDataset(x_train, y_train)
valid_dataloader = DataLoader(dataSet, batch_size=32, shuffle=True) # 배치크기 32, + Data shuffle

print(x_train)
print(y_train)

class DNN(nn.Module):
  def __init__(self, x_len, n_classes):
    super(DNN, self).__init__()
    self.layer1 = nn.Linear(x_len, 10)
    self.layer2 = nn.Linear(10, 10)
    self.layer3 = nn.Linear(10, n_classes)
    self.activate_func = nn.ReLU()

  def forward(self, x):
    x = self.layer1(x)
    x = self.activate_func(x)
    x = self.layer2(x)
    x = self.activate_func(x)
    logit = self.layer3(x)
    #x = self.atFunc(x)
    #logit = self.l3(x)
    return logit

data = []

model = DNN(x_len, 7).to(DEVICE) # input은 x의 속성수, output은 classification(7가지 error들)

optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
nb_epochs = 1000

for epoch in range(nb_epochs + 1):
  #Training Part
  for batch_idx, loaddata in enumerate(dataloader):
    x_input, y_input = loaddata
    logit = model(x_input)
    y_input = y_input.view(-1) # cross_entropy 사용을 위한 크기 맞춰줌

    loss = F.cross_entropy(logit, y_input)
    optimizer.zero_grad()
    loss.backward()
    try:
      optimizer.step()
    except:
      print(logit.shape)
      print(y_input.shape)

    if batch_idx % 3 == 0:
     print("epoch : %d, step : %d, loss : %s"%(epoch, batch_idx, loss.data))
     #data.append(loss.data.cpu().detach().numpy())
  
  correct = 0
  total_size = 0


  # test part (확인을 위해 Training과 같은 데이터를 사용하였음.)
  for batch_idx, loaddata in enumerate(dataloader2):
    x_input, y_input = loaddata
    logit = model(x_input)
    y_input = y_input.view(-1)

    # 값을 최대로갖는 index와 정답 index가 같은경우만 correct 1 증가
    # 각 output에 대해서 정답 index 추출 
    logit = torch.argmax(logit, dim = 1).to(DEVICE) 
    q = (logit == y_input).type(torch.LongTensor).to(DEVICE)

    correct += q.sum() # 정답의 수 더해줌
    total_size += len(x_input) # 전체 문제의 수

  print("***************************************")
  print(round((correct.data.cpu().detach().numpy() / total_size) * 100, 2), "% accuracy")
  print("***************************************")
```
